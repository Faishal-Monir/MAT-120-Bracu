<h3><b> 4) Sympy Lambdify </b><br>-->To remove the restriction we can use lambdify()<br>-->you will learn more in this cell !!<h4>Read this note for better understanding !</h4>

<a href="https://www.tutorialspoint.com/sympy/sympy_lambdify_function.htm">More about this topic !</a>

````python
a=sin(x)+cos(y)
l=lambdify([x,y],a) #lambdify maps the values as [x,y] that will take the values and--> ,a that will map the values in the a memory of the designated ram
l(30,60)            # calling l as a function passing the values of x-->30,y-->60 . Note: the values are in radian format thats why the results are a bit weird
````
