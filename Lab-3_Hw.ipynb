```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
```

```python
a,b,x=smp.symbols("a,b,x")
f=smp.sin(a*x)*smp.exp(-b*x)
w=smp.Integral(f,(x,0,1))
display(w)
```

```python
d=((a-smp.cos(x))**2)+((b-smp.sin(x))**2)
g=1/d
display(smp.Integral(g,(x,0,2*smp.pi)))
print("\n")
display(smp.integrate(d,(x,0,smp.pi)))
var=smp.integrate(d,(x,0,np.pi))
display(var)
print("\n")
display(var.evalf(3))
```
