```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def trapezoidal_rule(f,a,b,n):
    f=smp.lambdify(x,f)
    h=(b-a)/n
    sum=(f(a)+f(b))/2
    for i in range (1,n):
        x_i=a+i*h
        sum+=f(x_i)
    return (sum*h)
x=smp.symbols('x')
f=x**2
a=0
b=10
n=100
print(smp.integrate(f,(x,a,b)).evalf())
print(trapezoidal_rule(f,a,b,n))
```

```python
y=smp.symbols("y")
expr=x**2-y
sol,solm=smp.solve(expr,x)
display(sol,solm)
```

$dy/dx =-y^2/x^2$

$Initial Condition: y(1)=1/2$

```python
y=smp.Function('y')
dy=smp.diff(y(x),x)
display(y(x))
display(dy)

test=dy+(y(x)**2)/x**2
display(test)

sol1=smp.dsolve(test,y(x),ics=())
sol2=smp.dsolve(test,y(x),ics={y(1): 1/2})
print("="*60)
print("Solution without initial condition:")
display(sol1)
print("Solution with initial condition:")
display(sol2)
```
