<a href="https://www.youtube.com/watch?v=Rn9Gr52zhrY" target="Blank"><b>More about trapezoidal rule ! </b></a>


Screenshot 2023-08-02 205542.png

```python
#The formula of trapezoidal rule !
#dell(x)= (upper limit- lower limit)/n --> where n is the number of intervals
#Tb= [(dell(x))/2] *[f(0)+2f(1)+2f(2)+2f(3)+2f(4)....+2f(n-1)+f(n)]
```

```python
"""
f=function
l=lower limit
u= Upper limit
n=number of parts
"""

def trapz(f,l,u,n):

    exprn=smp.lambdify(x,f)

    dell=(u-l)/n
    summation=0.5*(exprn(l)+exprn(u))

    for i in range(1,n):
        data=l+(i*dell)
        summation+=exprn(data)

    return summation*dell


f=smp.sin(x)
display(smp.Integral(f,(x,0,smp.pi)))
pl()
#======================Trapezoidal rule call============================================
print("Trapezoidal rule approximation:\n")
h=trapz(f,0,np.pi,1000)
display(h)
#========================================================================================
pl()
print("The original ans:")
display(smp.integrate(f,(x,0,np.pi)).evalf()) #The acctual integration result produces by sympy integration and floating estimation
```
