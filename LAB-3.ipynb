```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=smp.symbols("x")
f=smp.sin(x)                   #just change here to see any trigonometri graphs cos,tan,csc,cot ,anything
print("f(x) = ",end=" ")
display(f)

d=np.arange(-np.pi, np.pi,0.45)
print(d)
print("\n")
flamb=smp.lambdify(x,f)
print(flamb(d))

print("\n")
plt.plot(d,flamb(d))
```

```python
df=smp.diff(f,x)
print("This is the ddx of sinx")
display(df)

df2=smp.diff(f,x,2)
print("This is the ddx of cos x")
display(df2)


b=np.arange(-np.pi, np.pi,0.45)
#print(b)
print("\n")
flam=smp.lambdify(x,df2)
#print(flam(b))

print("\n")
plt.plot(b,flam(b))
```

```python
#ddx of a polynomial

b=9*x**3-7*x**2+3*x+10
print("f(x)=")
display(b)
print("\n")
d=np.arange(-5,6,1)
print(d)
print("\n")

fun=smp.lambdify(x,b)
print(fun(b))
print("\n")
plt.plot(d,fun(d))

```

#INTEGRATION

```python
#Integration
fdx=smp.exp(-x)*(x**2)*smp.cos(x)
display(smp.Integral(fdx,(x,0,smp.ln(4))))
print("\n")
a=smp.integrate(fdx,(x,0,np.log(4)))
display(a)
print("\n")
display(smp.simplify(a))
```

```python
#Integration
fda=smp.exp(x)/(smp.sqrt(smp.exp(x)+9))
display(smp.Integral(fda,(x,0,smp.ln(4))))
print("\n")
b=smp.integrate(fda,(x,0,np.log(4)))
display(b)
print("\n")
display(smp.simplify(b))
```
