```python
from sympy import *
from numpy import *
from scipy.integrate import quad
#===================
import numpy as np
import sympy as smp
import math
import matplotlib.pyplot as plt

#Just a line printing function
def pl():
    print("\n","*"*100)
```

```python
x=smp.symbols("x")
```

```python
#Question1 Solve

fn=((smp.sin(x)*smp.cos(x))/(1-((1/5)*(x)**2)))
display(fn)
pl()

# #Derrivative
print("Derrivative Ans:")
temp=smp.diff(fn,x)
display(temp)
pl()

h=lambdify(x,fn)
dh=lambdify(x,temp) #for the derrivative function

r=np.arange(-5,5,0.1)  #change this later
# display(r)


p=h(r)
dp=dh(r)

#_________________________________________
plt.plot(r,p,c="blue")
plt.plot(r,dp,c="blue")
plt.axhline(y=0,c="Gray")
plt.axvline(x=0,c="Gray")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#a)
def trapezoidal_rule(f,l,u,n):

    exprn=smp.lambdify(x,f)

    dell=(u-l)/n
    summation=0.5*(exprn(l)+exprn(u))

    for i in range(1,n):
        data=l+(i*dell)
        summation+=exprn(data)

    return summation*dell


f=((smp.tan(x)**3)*ln(x))/x
display(smp.Integral(f,(x,1,50)))
pl()
#======================Trapezoidal rule call============================================
print("Trapezoidal rule approximation:\n")
h=trapezoidal_rule(f,1,50,1000)
display(h)
```

```python
#b)
fn=((smp.tan(x)**3)*ln(x))/x
f=smp.lambdify(x,fn)
l=1
u=50
result,error=quad(f,l,u)
print(f"Approx value of the integral is: {result}\nEstimate of the error: {error}")
```

```python
#c)
fn=((smp.tan(x)**3)*ln(x))/x
display(fn)
pl()

#integration
print("Printing the question in integral form:\n")
display(smp.Integral(fn,(x,1,50)))
pl()

print("Integrating Ans:\n")
temp=smp.integrate(fn,(x,1,50))
display(temp)
pl()
print("The floating value of the integration: \n")
display(smp.integrate(fn,(x,1,50)).evalf())            #used lower limit 1 as 0 gives a null value
```

```python
import math
def gamma(data):
    fn=(x**(data-1)*(smp.exp(-x)))
    display(fn)
    pl()
    display(smp.Integral(fn,(x,0,math.inf)))
    pl()
    print("Decimal Output:\n")
    temp=smp.integrate(fn,(x,0,math.inf))
    display(temp)
    pl()
    return temp
```
